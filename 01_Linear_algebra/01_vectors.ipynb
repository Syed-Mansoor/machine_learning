```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Distance from origin
a = np.array([1,2,3,4,5])
# calculate the euclidean distance from the origin (l2 norm)
distance = np.linalg.norm(a)

print("Euclidean Distance from the origin",distance)
```

```python
# to find euclidean distance in n-d
import numpy as np

# define two n-dimensional vectors a and b
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])

# calculate the difference vector
difference = a - b
# calculate the euclidean distance between a and b (l2 norm  of the differnecece)
distance = np.linalg.norm(difference)

print('Euclidean distance between a and b',distance)
```

```python
import numpy as np
import matplotlib.pyplot
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.distance import euclidean

# 1. generate 5 random 3D vectors
vectors = np.random.rand(5,3)

# 2. Assign a random class (0 or 1) to each vector
classes = np.random.randint(0,2,5)

# 3. plot the vectors in a 3D matplotlib graph
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
for vec,cls in zip(vectors,classes):
    ax.quiver(0,0,0,vec[0],vec[1],vec[2],color = {'r' if cls == 1 else "b"})

# get user input for a query point(3d vector)

query_vector = np.array(list(map(float,input('Enter the query point (3D vector) seperated by space :').split())))
# 5. Calculate the distance from the query vector to the 5 vectors and find the nearest neighbor
distances = [euclidean(query_vector,vec) for vec in vectors]
nearest_neighbor_index  = np.argmin(distances)

# 6. output the class of the nearest neighbor
print('The class of nearest neighbor is :',classes[nearest_neighbor_index])

# 7. plot hte query vector with a different color
ax.quiver(0,0,0,query_vector[0],query_vector[1],query_vector[2],color = 'g')
ax.set_xlabel('X-axis ')
ax.set_ylabel('Y_axis')
ax.set_zlabel('z_axis')
plt.show()
```

# Mean centering

```python
import numpy as np
import matplotlib.pyplot as plt

# generate a random 2D data with 100 points

data = np.random.rand(100,2)
data
```

```python
# plot the original data
plt.figure(figsize = (12,6))
plt.subplot(1,2,1)
plt.scatter(data[:,0],data[:,1],label = 'original_ data')
plt.xlabel("X-axis")
plt.ylabel('y_axis')
plt.title('Vefore Mean centering ')
plt.axhline(0,color  = 'black',linewidth = 0.5)
plt.axvline(0,color = 'black',linewidth = 0.5)
plt.legend()
```

```python
# calculate the man of the data along each dimension
data_mean = np.mean(data,axis =0)

print(data_mean)

# perform mean centering
centered_data = data - data_mean
print(centered_data)

# plot the mean centered data
plt.subplot(1,2,2)
plt.scatter(centered_data[:,0],centered_data[:,1],label = 'mean_centered_data',color = 'orange')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('After mean_centering')
plt.axhline(0,color = 'green',linewidth = 0.5)
plt.axvline(0,color = 'green',linewidth = 0.5)
plt.legend()
plt.show()
```

# Dot product

```python
# define the vectors
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dot(a,b))
print(a @ b)
```

```python
# define three vectors
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([7,8,9])

# calculate the dot product using np.dot() functons
dot_product_ab = np.dot(a,b)
dot_product_ba = np.dot(b,a)

print(dot_product_ab,dot_product_ba)

# check commutative property a * b = b * a
if dot_product_ab == dot_product_ba:
    print('The dot product is commutative')
else :
    print('The dot product is not commutative')
```

```python
# check distributive property a *(b + c) = a * b _ a * c
dot_product_a_bc = np.dot(a,b+ c)
dot_product_ab_plus_ac = np.dot(a,b) + np.dot(a,c)
print(dot_product_a_bc,dot_product_ab_plus_ac)

if dot_product_a_bc == dot_product_ab_plus_ac:
    print('the product is distributive')
else:
    print('The product is not distributive')
```

# cosine similarity(the cos of angle between two vectors)

```python
# define the vectors
a = np.array([1,2,3])
b = np.array([-4,-5,-6])
c = np.array([5,5,5])

# calculate the cosine similarity
cosine_similarity = np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
print('Cosine similarity between a and b :',cosine_similarity)

# calculate the cosine similarity
cosine_similarity = np.dot(a,c)/(np.linalg.norm(a) * np.linalg.norm(b))
print('Cosine similarity between a and c :',cosine_similarity)
```
