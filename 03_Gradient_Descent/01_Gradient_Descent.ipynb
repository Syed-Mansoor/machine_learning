```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets = 1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# lets apply ols
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red')
```

```python
# lets apply gradient descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
y_pred = ((78.35 * x) + 0).reshape(4)   # y = m*xi + b
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'ols')
plt.plot(x,y_pred,color = '#00a65a',label  = 'b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
loss_slope =  -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
# let's take learning rate = 0.01
lr = 0.1
```

```python
step_size = loss_slope * lr
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'ols')
plt.plot(x,y_pred1,color = '#00a65a',label  = 'b = {}'.format(b))
plt.plot(x,y_pred,color = '#A3E4D7',label = 'b = 0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2 * np.sum(m * x.ravel() - b)
print(loss_slope)

step_size = loss_slope * lr
print(step_size )
b = b - step_size
print(f'{b : }')
y_pred2 = ((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'ols')
plt.plot(x,y_pred2,color = '#00a65a',label  = 'b = {}'.format(b))
plt.plot(x,y_pred1,color = '#A3E4D7',label = 'b = {}'.format(b))
plt.plot(x,y_pred,color = '#A3E4D7',label = 'b = 0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slope = -2 * np.sum(y - m * x.ravel() - b)
print(f'{loss_slope : }')
step_size = loss_slope * lr
print(f'{step_size : }')
b = b - loss_slope
print(f'{b : }')
y_pred3 = ((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'ols')
plt.plot(x,y_pred3,color = '#00a65a',label = 'b = {}'.format(b))
plt.plot(x,y_pred2,color = '#A3E4D7',label = 'b = {}'.format(b))
plt.plot(x,y_pred1,color = '#A3E4D7',label  = 'b = {}'.format(b))
plt.plot(x,y_pred,color = '#A3E4D7',label = 'b = 0')
plt.legend()
plt.show()
```

```python
# iteration 4
loss_slope = -2 * np.sum(m * x.ravel() - b)
print(f'{loss_slope : }')
step_size = loss_slope * lr
print(f'{step_size : }')
b = b - step_size
print(f'{b : }')
y_pred4 = ((78.35 * x) + b) .reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'ols')
plt.plot(x,y_pred4,color = '#00a65a',label = 'b = {}'.format(b))
plt.plot(x,y_pred3,color = '#A3E4D7',label = 'b = {}'.format(b))
plt.plot(x,y_pred2,color = '#A3E4D7',label = 'b = {}'.format(b))
plt.plot(x,y_pred1,color = '#A3E4D7',label  = 'b = {}'.format(b))
plt.plot(x,y_pred,color = '#A3E4D7',label = 'b = 0')
plt.legend()
plt.show()
```

```python
# using loop
b = -100
m  = 78.35
lr = 0.1

epochs = 10
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m * x.ravel() - b )
    b = b - (lr * loss_slope)
    y_pred = m * x + b
    plt.plot(x,y_pred)
plt.scatter(x,y)
plt.show()
```
